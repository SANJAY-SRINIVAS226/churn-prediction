```python
import pandas as pd
import numpy as np

# Keeps random numbers the same
np.random.seed(42)

# 1. Load the original Kaggle file from the data folder
# The '../' tells Python to go up one folder, then look inside 'data'
df = pd.read_csv(r"C:\Users\Admin\churn-prediction\Dataset\WA_Fn-UseC_-Telco-Customer-Churn.csv")

# 2. Rename existing columns
df = df.rename(columns={
    'customerID': 'CustomerID',
    'gender': 'Gender',
    'tenure': 'TenureMonths'
})

num_rows = len(df)

# 3. Generate fake features
df['Age'] = np.random.randint(18, 70, size=num_rows)
df['SubscriptionType'] = np.random.choice(["Basic", "Premium", "Enterprise"], size=num_rows)
df['SupportTickets'] = np.random.randint(0, 11, size=num_rows)
df['SatisfactionScore'] = np.random.randint(1, 11, size=num_rows)
df['UsageHoursPerWeek'] = np.random.randint(1, 41, size=num_rows)
df['PaymentMethod'] = np.random.choice(["Credit Card", "UPI", "Net Banking", "Debit Card"], size=num_rows)

# 4. Filter for your required 11 columns
required_columns = [
    'CustomerID', 'Age', 'Gender', 'SubscriptionType', 'MonthlyCharges', 
    'TenureMonths', 'SupportTickets', 'PaymentMethod', 'SatisfactionScore', 
    'UsageHoursPerWeek', 'Churn'
]
final_df = df[required_columns]

# 5. Save the new custom dataset back into the data folder
final_df.to_csv(r"C:\Users\Admin\churn-prediction\Dataset\telco_with_fake_features.csv", index=False)

print(f"Done! Created dataset with {final_df.shape[0]} rows and {final_df.shape[1]} columns inside the data folder.")
```
