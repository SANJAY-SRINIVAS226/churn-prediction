```python
import pandas as pd
import numpy as np

np.random.seed(42)

# 1. Load the raw file
df_raw = pd.read_csv(r"C:\Users\Admin\churn-prediction\Dataset\WA_Fn-UseC_-Telco-Customer-Churn.csv")
num_rows = len(df_raw)
df = df_raw.copy()

# 2. Random features
df['Age'] = np.random.randint(18, 70, size=num_rows)
df['UsageHoursPerWeek'] = np.random.randint(10, 50, size=num_rows)
df['SubscriptionType'] = np.random.choice(["Basic", "Premium", "Enterprise"], size=num_rows)

# 3. REALISTIC BLENDED PATTERNS (Adding slight noise so it's not 100% perfect)
support_tickets = []
satisfaction_scores = []

for churn_status in df['Churn']:
    if churn_status == 'Yes':
        # 80% of churners have high tickets, 20% have low tickets
        tickets = np.random.randint(5, 11) if np.random.rand() < 0.8 else np.random.randint(0, 4)
        sat = np.random.randint(1, 4) if np.random.rand() < 0.8 else np.random.randint(5, 9)
    else:
        # 85% of loyal customers have low tickets, 15% have high tickets
        tickets = np.random.randint(0, 3) if np.random.rand() < 0.85 else np.random.randint(4, 8)
        sat = np.random.randint(6, 11) if np.random.rand() < 0.85 else np.random.randint(2, 5)
        
    support_tickets.append(tickets)
    satisfaction_scores.append(sat)

df['SupportTickets'] = support_tickets
df['SatisfactionScore'] = satisfaction_scores

# 4. Save file
df.to_csv(r"C:\Users\Admin\churn-prediction\Dataset\telco_with_all_features.csv", index=False)
print("🎯 Realistic, balanced data file created successfully!")
```
